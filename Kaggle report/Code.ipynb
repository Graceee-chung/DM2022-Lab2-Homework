#### >>> For each sentiment, select 500 most similar terms by word2vec_model as the whole features

```python
label = ['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust']
target_features = [i for i in list(word2vec_model.wv.most_similar(word, topn=500) for word in label)]
target_features = [j[0] for i in target_features for j in i]
```

```python
import numpy as np
from scipy.sparse import csr_matrix

vocabulary = {}
for term in target_features:
    vocabulary.setdefault(term, len(vocabulary))
print(len(vocabulary))

docs1 = [i.split() for i in X_train[:]]
docs2 = [i.split() for i in X_test[:]]
temp = [item for sublist in docs2 for item in sublist]
target = [] #[i for i in vocabulary if i not in temp]
for word in vocabulary:
    if word not in temp:
        target.append(word)

del temp
```

#### >>> Build csr_matrix based on target_features of text of train and test datasets

```python
import numpy as np
from scipy.sparse import csr_matrix

for w in target:
    vocabulary.pop(w, None)

print(len(vocabulary))

indptr1 = [0]
indices1 = []
data1 = []
indptr2 = [0]
indices2 = []
data2 = []

for d in docs1:
    for term in d:
        if term in vocabulary:
            index = vocabulary.get(term)
            indices1.append(index)
            data1.append(1)
        else:
            continue

    indptr1.append(len(indices1))

for d in docs2:
    for term in d:
        if term in vocabulary:
            index = vocabulary.get(term)
            indices2.append(index)
            data2.append(1)
        else:
            continue
      
    indptr2.append(len(indices2))

del docs1, docs2, vocabulary

X_train_cm = csr_matrix((data1, indices1, indptr1), dtype=int)
X_test_cm = csr_matrix((data2, indices2, indptr2), dtype=int)

del data1, data2, indices1, indices2, indptr1, indptr2
print('X_train.shape: ', X_train_cm.shape)
print('y_train.shape: ', X_train.shape)
print('X_test.shape: ', X_test_cm.shape)
print('y_test.shape: ', X_test.shape)
```

```python
input_shape = X_train_cm.shape[1]
output_shape = 8
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers import ReLU, Softmax

# input layer
model_input = Input(shape=(input_shape, ))  # 500
X = model_input

# 1st hidden layer
X_W1 = Dense(units=100)(X)  # 64
H1 = ReLU()(X_W1)

# 2nd hidden layer
H1_W2 = Dense(units=60)(H1)  # 64
H2 = ReLU()(H1_W2)

# output layer
H2_W3 = Dense(units=output_shape)(H2)  # 4
H3 = Softmax()(H2_W3)

model_output = H3

# create model
model = Model(inputs=[model_input], outputs=[model_output])

# loss function & optimizer
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# show model construction
model.summary()
```
